```python
import re 

class TiendaOnline:
    def __init__(self):
        self.inventario = []
        self.clientes = {}
        self.ventas_totales = 0

        
    def agregar_producto(self, nombre, precio, cantidad):
            for producto in self.inventario:
                if producto["nombre"] == nombre:
                        producto["cantidad"] += cantidad
                        break
            else:
                self.inventario.append({"nombre": nombre, "precio": precio, "cantidad": cantidad})
    def ver_inventario(self):
        for producto in self.inventario:
            print(f"Nombre: {producto['nombre']}, Precio: {producto['precio']}, Cantidad: {producto['cantidad']}")

    def buscar_producto(self, nombre):
        for producto in self.inventario:
            if producto["nombre"].lower() == nombre.lower():
                return f"Nombre: {producto['nombre']}, Precio: {producto['precio']}, Cantidad: {producto['cantidad']}"
        return "Producto no encontrado"

    def actualizar_stock(self, nombre, cantidad):
            for producto in self.inventario:
                if producto["nombre"].lower() == nombre.lower():
                    if producto["cantidad"] > cantidad:
                        print("no se puede, no hay suficiente")
                    else:
                        producto["cantidad"] += cantidad
                    break
            else:
                print("Producto no encontrado")
    def eliminar_producto(self, nombre):
        for producto in self.inventario:
            if producto["nombre"].lower() == nombre.lower():
                self.inventario.remove(producto)
                print(f"Producto {nombre} eliminado.")
                break
        else:
            print("Producto no encontrado")  
            
    def valor_inventario(self):
        valor_total = 0
        for producto in self.inventario:
            valor_total += producto["precio"] * producto["cantidad"] 
        return valor_total
    
    def buscar_producto_regex(self, patron):
      
        coincidencia_productos = []
        patron = r'[a-zA-Z]+'
        
        for i in self.inventario:
            i = re.findall(patron, coincidencia_productos['nombre'])
            if len(coincidencia_productos) >=1:
                coincidencia_productos.append(i)
        print(coincidencia_productos)  
            
    def realizar_compra(self):
        self.carrito = []
        while True:
            print("\nInventario actual:")
            self.ver_inventario()
            nombre_producto = input("Ingrese el nombre del producto que desea comprar (o 'terminar' para finalizar la compra): ")
            if nombre_producto.lower() == 'terminar':
                break
            encontrado = False
            for producto in self.inventario:
                if producto["nombre"].lower() == nombre_producto.lower():
                    encontrado = True
                    if producto["cantidad"] > 0:
                        self.carrito.append(producto)
                        producto["cantidad"] -= 1
                        print(f"Producto {producto['nombre']} agregado al carrito.")
                        print("Contenido del carrito:")
                        for item in self.carrito:
                            print(f"Nombre: {item['nombre']}, Precio: {item['precio']}")
                    else:
                        print("El producto está agotado.")
                    break
            if not encontrado:
                print("Producto no encontrado en el inventario.")
        total_compra = sum(producto["precio"] for producto in self.carrito)
        print(f"\nProductos en el carrito:")
        for producto in self.carrito:
            print(f"Nombre: {producto['nombre']}, Precio: {producto['precio']}")
        print(f"Total de la compra: {total_compra}")
        self.procesar_pago(total_compra)
        
    def procesar_pago(self, total_compra):
        while True:
            try:
                cantidad_a_pagar = float(input("Ingrese la cantidad a pagar: "))
                cantidad_pagada = float(input("Ingrese la cantidad pagada: "))
                if cantidad_a_pagar != total_compra:
                    print("Esa no es la cantidad correcta.")
                elif cantidad_pagada < total_compra:
                    print("Cantidad insuficiente. Intente de nuevo.")
                else:
                    cambio = cantidad_pagada - total_compra
                    print(f"Cambio: {cambio}")
                    print("Pago procesado correctamente.")
                    break
            except ValueError:
                print("Por favor, ingrese valores numéricos.")        

    def agregar_cliente(self, nombre, email):
        if nombre not in self.clientes:
            self.clientes[nombre] = {"email": email, "historial_compras": []}
            print(f"Cliente {nombre} agregado correctamente.")
        else:
            print(f"El cliente {nombre} ya existe.")
            
    def ver_clientes(self):
        if self.clientes:
            print("Detalles de los clientes:")
            for cliente, detalles in self.clientes.items():
                print(f"Nombre: {cliente}, Email: {detalles['email']}, Compras realizadas: {detalles['historial_compras']}")
        else:
            print("No hay clientes registrados.")
      
    def registrar_compra(self, nombre_cliente, carrito):
        if nombre_cliente in self.clientes:
            total_compra = sum(producto["precio"] * producto["cantidad"] for producto in carrito.values())
            self.ventas_totales += total_compra
            self.clientes[nombre_cliente]["historial_compras"].append({"productos": carrito, "total": total_compra})
            print(f"Compra registrada para {nombre_cliente} por un total de {total_compra}.")
        else:
            print(f"Error: el cliente {nombre_cliente} no está registrado.") 
    def ver_compras_cliente(self, nombre_cliente):
        if nombre_cliente in self.clientes:
            compras = self.clientes[nombre_cliente]["historial_compras"]
            if compras:
                print(f"Detalles de las compras del cliente {nombre_cliente}:")
                for i, compra in enumerate(compras, start=1):
                    print(f"Compra {i}:")
                    for producto, detalles in compra["productos"].items():
                        print(f"Producto: {producto}, Precio: {detalles['precio']}, Cantidad: {detalles['cantidad']}")
                    print(f"Total de la compra: {compra['total']}")
                    print("------------")
            else:
                print(f"El cliente {nombre_cliente} no ha realizado compras.")
        else:
            print(f"Error: el cliente {nombre_cliente} no está registrado.") 
            
    def calcular_ventas_totales(self):
        print(f"Las ventas totales de la tienda son: {self.ventas_totales}")        
```

```python
# Ejemplo de uso
tienda = TiendaOnline()

tienda.agregar_producto("camisa", 10, 5)
tienda.agregar_producto("pantalón", 10, 5)
tienda.agregar_producto("camiseta", 10, 5)
tienda.agregar_producto("zapatos", 10, 5)


tienda.ver_inventario()

tienda.buscar_producto('cucharas')


tienda.agregar_cliente("Cliente1", "cliente1@example.com")
tienda.agregar_cliente("Cliente2", "cliente2@example.com")

carrito_cliente = {"camisa": {"precio": 10,"cantidad":3},"calcetines":{"precio":7,"cantidad":2}}
tienda.registrar_compra("Cliente1", carrito_cliente)
carrito_cliente = {"camisa": {"precio": 10,"cantidad":3}}
tienda.registrar_compra("Cliente1", carrito_cliente)
carrito_cliente = {"camisa": {"precio": 10,"cantidad":3}}
tienda.registrar_compra("Cliente1", carrito_cliente)
carrito_cliente = {"jamon": {"precio": 10,"cantidad":3},"chope":{"precio":7,"cantidad":2}}
tienda.registrar_compra("Cliente2", carrito_cliente)


tienda.procesar_pago()

tienda.ver_compras_cliente("Cliente5")
tienda.ver_clientes()
tienda.calcular_ventas_totales()
```
